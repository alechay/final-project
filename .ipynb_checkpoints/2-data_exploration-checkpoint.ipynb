## Wrote functions to load the data

Read one inertial signals file into a pd dataframe. Training data is composed of 7,352 rows or windows of data, where each window has 128 observations. This means that for all subjects in the training set, there was a total of 7,352 windows of data collected. Each window has 128 observations because they were 2.56 seconds and there were 50 data points per second. (2.56 * 50 = 128)

```python
filepath = 'UCI HAR Dataset/train/Inertial Signals/total_acc_y_train.txt'
df = pd.read_csv(filepath, header=None, delim_whitespace=True)
df.values.shape
```

Wrote a function to do what has been done above. The function loads a single inertial signal txt file as a numpy array. The same filepath was used and you'll notice it has the same shape as the one above

```python
# load a single txt file as a numpy array
def load_file(filepath):
    df = pd.read_csv(filepath, header=None, delim_whitespace=True)
    return df.values

load_file(filepath).shape
```

Wrote a function to load a group of variables into a 3D numpy array. For example, training data observation windows for total_acc in the x, y, and z dimensions are stacked to return a numpy array with dimensions (7352, 128, 3).

```python
# load a group of files, such as x y and z data for a given variable
def load_group(filenames, directory=''):
    loaded = []
    for name in filenames:
        data = load_file(directory + name)
        loaded.append(data)
    # stack group so that features are in the 3rd dimension
    loaded = np.dstack(loaded)
    return loaded

train_dir = 'UCI HAR Dataset/train/Inertial Signals/'
filenames = ['total_acc_x_train.txt', 'total_acc_y_train.txt', 'total_acc_z_train.txt']
total_acc = load_group(filenames, directory=train_dir)
total_acc.shape
```

Now wrote a function to load all inertial signals in a group, train or test, as a 3D numpy array. For example, training data observation windows for total_acc, body_acc, body_gyro, in the x, y, and z dimensions are stacked to return a numpy array with dimensions (7352, 128, 9). This function also loads the corresponding output values for the train or test group as a numpy array. Therefore, it creates a 1D array of activity labels (1:walking, 2:walking_upstairs, 3:walking_downstairs, 4:sitting, 5:standing, 6:laying) that is just as long as the number of windows that were collected (7352, 1). 

```python
# load a dataset group, such as train or test
def load_dataset(group, main_data_dir=''):
    filepath = main_data_dir + group + '/Inertial Signals/'
    # get all 9 files in Inertial Signals into a single filename list
    filenames = []
    # total acceleration
    filenames += ['total_acc_x_'+group+'.txt', 'total_acc_y_'+group+'.txt', 'total_acc_z_'+group+'.txt']
    # body acceleration
    filenames += ['body_acc_x_'+group+'.txt', 'body_acc_y_'+group+'.txt', 'body_acc_z_'+group+'.txt']
    # body gyroscope
    filenames += ['body_gyro_x_'+group+'.txt', 'body_gyro_y_'+group+'.txt', 'body_gyro_z_'+group+'.txt']
    # load input data
    X = load_group(filenames, filepath)
    # load class output 
    y = load_file(main_data_dir + group + '/y_'+group+'.txt')
    return X, y

# load train data
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
print("train:", trainX.shape, trainy.shape)

# load test data
testX, testy = load_dataset('test', 'UCI HAR Dataset/')
print("test:", testX.shape, testy.shape)
```

## Check the breakdown of the data

Now we want to see that we had a similar amount of observation windows for each activity, and whether the breakdown was similar for the test and train datasets. Therefore, I looked at the number of occurances of each value in the output dataset, and I found that the dataset was balanced.

```python
# summarize the balance of class in an output variable column
def class_breakdown(data):
    # convert numpy array to dataframe
    df = pd.DataFrame(data)
    # get value_counts number
    counts = df[0].value_counts()
    # get value_counts percentage
    percent = df[0].value_counts(normalize=True)*100
    # summarize
    for i in np.sort(df[0].unique()):
        print('Activity=%d, total=%d, percentage=%.3f' % (i, counts[i], percent[i]))
        
# train file class breakdown
print(color.BOLD + 'Train Dataset:' + color.END)
class_breakdown(trainy)

#test file class breakdown
print(color.BOLD + 'Test Dataset:' + color.END)
class_breakdown(testy)

# summarize combined class breakdown
print(color.BOLD + 'Both:' + color.END)
combined = np.vstack((trainy, testy))
class_breakdown(combined)
```

## Get sequential movement data for one subject

Now that we have the indeces that belong to a specific subject, we can use that to get the observation windows (X) and activity id per observation window lists (y) for each subject. Wrote a function to do just that. The output of the below cell shows the X and y data shape for subject #1

```python
# get all data for one subject
def data_for_subject(X, y, sub_map, sub_id):
    # get row indexes for the subject id
    ix = [i for i in range(len(sub_map)) if sub_map[i]==sub_id]
    # return the selected samples
    return X[ix, :, :], y[ix]

# load train data
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
# load subject id per observation window file
sub_map = load_file('UCI HAR Dataset/train/subject_train.txt')
# get sub_id number 1
sub_id = train_subjects[0]
# get the data for one subject
subX, suby = data_for_subject(trainX, trainy, sub_map, sub_id)
print("Subject", sub_id, "X data shape:", subX.shape)
print("Subject", sub_id, "y data shape:", suby.shape)
```

The observation windows had 50% overlap with each other, per the README file. We need to get rid of this overlap to get a sequential list of inertia signals. Wrote a function to do that:

```python
# convert a series of windows to a 1D list
def to_series(windows):
    series = []
    for window in windows:
        # remove the overlap from the window
        half = int(len(window) / 2)
        for value in window[-half:]:
            series.append(value)
    return series
```

subX[:,:,0] is the total_acc in the x dimension. Using this function creates a sequential series of inertia measurements, rather than 128 sets of overlapping observation windows. Therefore, the dimensionality is reduced: (347, 128) -> (22208, 1).

Wrote a function to create plots for all 9 inertial txt files for one subject and the activity label. This will plot the sequential series of inertia observations for total_acc, body_acc, and body_gyro for the x, y, and z dimensions in subplots that share the x-axis. For all 9 plots, the x range should be 0-22208. It also plots the label of the activity being performed, and because this plot is on bottom, you see its x-axis with range 0-347. It hides the x-axes of the 9 txt file plots. When plotted on top of each other, these x-axes align because they are related to one another. 0-22208 is scaled by observation (there are now 64 observations per window and 22208 / 64 = 347) and 347 is scaled by observation window.

```python
# plot the data for one subject
def plot_subject(X, y):
    plt.figure(figsize=(10,8))
    # determine the total number of plots
    n, off = X.shape[2] + 1, 0
    # plot total acc
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('total acc '+str(i), y=0, loc='left')
        off += 1
    # plot body acc
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('body acc '+str(i), y=0, loc='left')
        off += 1
    # plot body gyro
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('body gyro '+str(i), y=0, loc='left')
        off += 1
    # plot activities
    plt.subplot(n, 1, n)
    plt.plot(y)
    plt.title('activity', y=0, loc='left')
    plt.show()
```

## Get movement data per activity

Now we want to discriminate between activities based on activity data. In the simplest form that means discriminating between activities in a single subject. In a single subject I will map the activity to its corresponding inertial data.

```python
# load data
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
# load mapping of rows to subjects
sub_map = load_file('UCI HAR Dataset/train/subject_train.txt')
train_subjects = np.unique(sub_map)
# get the data for one subject
sub_id = train_subjects[0]
subX, suby = data_for_subject(trainX, trainy, sub_map, sub_id)
print(subX.shape)
print(suby.shape)
```

Define a function that gets the data based on the activity the subject is performing. Next 3 cells are confirm behavior of the function, and they are raw to mute their output

```python
# group data by activity
def data_by_activity(X, y, activities):
    # group windows by activity
    return {a:X[y[:,0]==a, :, :] for a in activities}
```

For one subject, write function to plot histogram of total_acc x, y, z, for each activity

```python
# plot total_acc histograms for each activity for a subject
def plot_activity_total_acc_hist(X, y):
    # get a list of unique activities for the subject
    activity_ids = np.unique(y[:,0])
    # group windows by activity
    grouped = data_by_activity(X, y, activity_ids)
    # plot per activity, histograms for each axis
    fig, axs = plt.subplots(nrows=6, ncols=1, sharex=True, figsize=(7.5,10))
    for k in range(len(activity_ids)):
        act_id = activity_ids[k]
        # total acceleration
        for i in range(3):
            data = to_series(grouped[act_id][:,:,i])
            axs[k].hist(data, bins=100)
            axs[k].set_title('activity ' +str(act_id), y=0, loc='left')
    plt.show()
```

## Confirm that activity classes are balanced

Now we want to confirm that the activities for a given subject over the course of their trace are balanced. We will see how many observation windows each subject spends on each activity and look at the distribution of durations of activity across subjects.

```python
# load data
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
# load mapping of rows to subjects
sub_map = load_file('UCI HAR Dataset/train/subject_train.txt')
train_subjects = np.unique(sub_map)
# get the data for one subject
sub_id = train_subjects[0]
_, subj_y = data_for_subject(trainX, trainy, sub_map, sub_id)
```

```python
len(subj_y[subj_y[:,0]==1])
```

Create function that plots the activity durations across subjects

```python
# plot activity durations by subject
def plot_activity_durations_by_subject(X, y, sub_map):
    # get unique subjects and activities
    subject_ids = np.unique(sub_map[:,0])
    activity_ids = np.unique(y[:,0])
    # enumerate subjects
    activity_windows = {a:[] for a in activity_ids}
    for sub_id in subject_ids:
        # get data for one subject
        _, subj_y = data_for_subject(X, y, sub_map, sub_id)
        # count windows by activity
        for a in activity_ids:
            activity_windows[a].append(len(subj_y[subj_y[:,0]==a]))
    durations = [activity_windows[a] for a in activity_ids]
    plt.boxplot(durations, labels=activity_ids)
    plt.show()
```

Plot activity durations for training data

```python
# load training data
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
# load mapping of rows to subjects
sub_map = load_file('UCI HAR Dataset/train/subject_train.txt')
# plot durations
plot_activity_durations_by_subject(trainX, trainy, sub_map)
```

Plot activity durations for test data

```python
# load test data
testX, testy = load_dataset('test', 'UCI HAR Dataset/')
# load mapping of rows to subjects
sub_map = load_file('UCI HAR Dataset/test/subject_test.txt')
# plot durations
plot_activity_durations_by_subject(testX, testy, sub_map)
```
